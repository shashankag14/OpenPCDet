```python
import pickle
import matplotlib.pyplot as plt
```

```python
pickle_file_path = '/mnt/data/shag01/3diou_proto_mcl/output/kitti_models/pv_rcnn_ssl_60/instance_idx.pkl'

with open(pickle_file_path, 'rb') as pickle_file:
    instance_data_dict = pickle.load(pickle_file)
```

```python
unique_instance_ids = list(instance_data_dict.keys())
id_to_index = {instance_id: index for index, instance_id in enumerate(unique_instance_ids)}
indices = [id_to_index[instance_id] for instance_id in unique_instance_ids]
counter_values = list(instance_data_dict.values())
print(len(unique_instance_ids))
```

```python
len(set(unique_instance_ids))
```

```python
counter_values = [data['Counter'] for data in instance_data_dict.values()]
class_labels = [data['Class'] for data in instance_data_dict.values()]
```

```python
class_label_color_map = {
    1: 'blue',
    2: 'green',
    3: 'red',
}
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(indices, counter_values, color=[class_label_color_map[class_label] for class_label in class_labels])
plt.xlabel('Unique Instances')
plt.ylabel('Counter Values')
plt.title('Counter Values for Unique Instances')
# plt.xticks(indices, fontsize=8)  # Adjust fontsize as needed
# plt.yticks(range(min(counter_values), max(counter_values) + 1))

# Create a legend for class labels and colors
class_labels_unique = set(class_labels)
handles = [plt.Line2D([0], [0], marker='o', color='w', label=class_label, markersize=10,
                      markerfacecolor=class_label_color_map[class_label]) for class_label in class_labels_unique]
plt.legend(handles=handles, title='Class Labels')

plt.show()
```
